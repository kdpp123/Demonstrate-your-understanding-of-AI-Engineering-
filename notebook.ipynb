### **DON'T LET YOUR MODEL CHEAT: A STORYTELLING APPROACH TO DATA SPLITTING!**

“Hey, do you have a minute?” my teammate Sana asked. “I’m trying to understand why we split data into training, validation, and test sets in machine learning. What’s the big deal?”

"Great question,” I smiled. “Let me tell you about a mistake I made early on — back when I completely ignored that step, even while training to become an AI Engineer."

I built a model to predict which customers might leave our service. I trained it on all our past data and got amazing results—super high accuracy. I was thrilled... until I tried using it on new customers. It totally flopped!

I explained, “The problem was, I trained and tested on the same data. The model had just memorized patterns instead of learning to generalize.”

Then I broke it down:
  
**Training data** teaches the model.
  
**Validation data** helps adjust and improve it during development.
  
**Test data** checks how it performs on completely new, unseen inputs.

“This split should happen before training,” I added. “Otherwise, it’s like preparing for a quiz by reading the answer key—sure, you’ll score well, but it doesn’t mean you actually understand the material.”

I laughed and said, “That model looked like a genius in practice, but in reality? It had stage fright in front of real customers.”

“That’s why data splitting is such an important step,” I added. “Without it, you risk building something that just pretends to be smart.”

Once I split the data properly, the model’s performance was more realistic. Not perfect—but honest.

Sana nodded. “So testing on new data is like giving your model a real-world quiz.”

“Exactly,” I said. “Without splitting, you don’t know if your model’s truly smart—or just really good at cramming.”

She smiled. "Got it. Thanks for the lesson!”
